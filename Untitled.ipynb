```python
df_dados["faixa etária"] = df_dados["age"].apply(lambda x: "até 30 anos" if x <= 30 else("De 31 anos a 40 anos" if x <= 40 else("De 41 anos a 50 anos" if x <= 50 else("De 51 anos a 60 anos" if x <= 60 else "Acima de 60 anos"))))
```

```python
df_dados
```

```python
df_dados["Regiões"] = df_dados["region"].apply(lambda x: "Noroeste" if x == 1 else(
                    "Nordeste" if x == 0 else(
                    "Sudeste" if x == 2 else(
                    "Sudoeste"))))
```

```python
df_dados
```

```python
clientes_idade = df_dados[['faixa etária','Regiões']].groupby('faixa etária', as_index=False).count()
clientes_regiao = df_dados[['Regiões','faixa etária']].groupby('Regiões', as_index=False).count()
```

```python
alt.Chart(clientes_idade, title="Clientes por faixa etária").mark_bar().encode(
    x= alt.X('faixa etária', title='Faixa etária'),
    y= alt.Y('Regiões', title='Quantidade de clientes'),
    color = alt.Color('faixa etária', legend=None)
).interactive()
```

```python
df_dados["Nordeste"] = df_dados["region"].apply(lambda x: "Sim" if x == 0 else "Não")
df_dados["Noroeste"] = df_dados["region"].apply(lambda x: "Sim" if x == 1 else "Não")
df_dados["Sudeste"] = df_dados["region"].apply(lambda x: "Sim" if x == 2 else "Não")
df_dados["Sudoeste"] = df_dados["region"].apply(lambda x: "Sim" if x == 3 else "Não")
```

```python
df_dados_sudeste = df_dados.loc[df_dados['Regiões'] == "Sudeste"]
df_dados_sudeste
```

```python
df_dados
```

```python
clientes_sudeste = df_dados_sudeste.groupby('faixa etária', as_index=False).count()[['faixa etária','Sudeste']]
clientes_sudeste
```

```python
alt.Chart(clientes_sudeste, title="Clientes por faixa etária no Sudeste").mark_bar().encode(
    x= alt.X('faixa etária', title='Faixa etária'),
    y= alt.Y('Sudeste', title='Quantidade de clientes'),
    color = alt.Color('faixa etária', legend=None)
).interactive()
```

```python
df_dados_sudoeste = df_dados.loc[df_dados['Regiões'] == "Sudoeste"]
df_dados_nordeste = df_dados.loc[df_dados['Regiões'] == "Nordeste"]
df_dados_noroeste = df_dados.loc[df_dados['Regiões'] == "Noroeste"]
```
